```python
import pandas as pd
import numpy as np
from matplotlib import pylab as pl
import sklearn 
%pylab inline
#from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import cross_val_score
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train
```

```python
Y_train = df_train['tare']
Y_train
```

```python
X_train = df_train['name']
X_train
```

```python
data_in_numpy = df_train['name'].values
data_in_numpy
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(data_in_numpy)
X_train_counts.shape
```

```python
X_train_counts
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_counts[:10000], Y_train[:10000])
# from sklearn.linear_model import SGDClassifier
# clf = SGDClassifier().fit(X_train_counts[:10000], Y_train[:10000])
```

```python
import numpy as np
predicted = clf.predict(X_train_counts[10000:])
np.mean(predicted == Y_train[10000:]) 
```
